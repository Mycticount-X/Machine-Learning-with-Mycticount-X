# Least Square Regression - Pseudo Inverse

`````python
import numpy as np
import matplotlib.pyplot as plt

# Data sample
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1])

# Buat matrix X (dengan kolom 1 untuk bias/intercept)
X = np.vstack([x, np.ones(len(x))]).T

# Hitung pseudo-inverse dari X
X_pseudo_inv = np.linalg.pinv(X)

# Hitung beta (koefisien regresi)
beta = X_pseudo_inv @ y

# Ekstrak slope (m) dan intercept (b)
m, b = beta[0], beta[1]

# Prediksi y menggunakan model regresi
y_pred = X @ beta

# Tampilkan hasil
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# Visualisasi
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Regresi Linear')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Least Squares Regression (Pseudo-Inverse)')
plt.grid(True)
plt.show()
`````

🧠 **Pseudocode**

```plaintext
INPUT:
  - x: array berisi data input (fitur)
  - y: array berisi data output (target)

LANGKAH-LANGKAH:

1. Buat matrix X dengan menambahkan kolom 1 sebagai bias/intercept:
      X ← matriks dengan baris [x_i, 1] untuk setiap elemen x_i

2. Hitung Pseudo-Inverse dari X:
      X_pseudo ← pseudo-inverse(X)

3. Hitung parameter regresi (β):
      β ← X_pseudo ⋅ y

4. Ekstrak slope dan intercept dari β:
      m ← β[0]      // slope
      b ← β[1]      // intercept

5. Prediksi nilai y menggunakan model regresi:
      y_pred ← X ⋅ β

6. (Opsional) Visualisasikan hasil:
      - Tampilkan titik data asli (x, y)
      - Tampilkan garis regresi (x, y_pred)

OUTPUT:
  - Nilai slope (m) dan intercept (b)
  - Plot regresi linear
````

---

🧮 **Rumus Matematika**

$\beta = (X^+) \cdot y$

Di mana $X^+$ adalah pseudo-inverse dari matriks $X$, dan $\beta = \begin{bmatrix} m \\ b \end{bmatrix}$

---

✅ **Kelebihan**

* Lebih stabil untuk sistem underdetermined atau ill-conditioned
* Tidak perlu menghitung $(X^T X)^{-1}$

⚠️ **Catatan**

* Pastikan input `x` dan `y` dalam bentuk vektor 1 dimensi dan memiliki panjang yang sama.
