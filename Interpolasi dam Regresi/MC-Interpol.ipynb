# Linear Interpolate

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# 1. Titik data contoh
xi = np.array([1, 3, 5])
yi = np.array([20, 45, 55])

# 2. Buat fungsi interpolasi linier
linear_interp = interp1d(xi, yi, kind='linear')

# 3. Buat domain uji (lebih rapat)
xtest = np.linspace(xi.min(), xi.max(), 100)

# 4. Hitung nilai interpolasi
ytest = linear_interp(xtest)

# 5. Visualisasi
plt.scatter(xi, yi, color='blue', label='Data Asli')
plt.plot(xtest, ytest, color='red', label='Interpolasi Linier')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Interpolation')
plt.legend()
plt.grid(True)
plt.show()
```

# Cubic Spilne Interpolate

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# 1. Data titik (x, y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 0.8, 0.9, 0.1, -0.8, -1])

# 2. Membuat interpolasi cubic spline
cs = CubicSpline(x, y)

# 3. Membuat titik x untuk interpolasi halus
x_dense = np.linspace(min(x), max(x), 500)
y_dense = cs(x_dense)

# 4. Plot hasil interpolasi
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'o', label='Data Asli')
plt.plot(x_dense, y_dense, label='Interpolasi Cubic Spline', color='red')
plt.title("Cubic Spline Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
