## Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Memberikan angka 1-10 dengan sebanyak 100 angka secara merata
x = np.linspace(0, 10, 100)

# Randomizer 
# np.random.random(len(x)) = Random Array 
y = 1 + x + x * 23 * np.random.random(len(x))


plt.scatter(x, y)
plt.show()
```

## Least Square Regresion

```python
x_vstack = np.vstack([x, np.ones(len(x))])
```

```python
x_final = x_vstack.T
```

```python
# Supaya bisa di transpose
y_final = y[:, np.newaxis]
```

```python
# beta = (xT . x)^(-1) . xT . y
beta = np.linalg.inv(x_final.T @ x_final) @ (x_final.T @ y_final)
beta
```

```python
m, b = beta

plt.scatter(x, y)
plt.plot(x, m * x + b, 'r-')
plt.title("Least Square Regresion: Data Random")
plt.show()
```

```python
# Variasi Pseudo Invers (Memaksa Invers Matrix Bukan Persegi)
beta_var = np.linalg.pinv(x_final) @ y_final
beta_var
```

```python
mp, bp = beta_var

plt.scatter(x, y)
plt.plot(x, mp * x + bp, 'r-')
plt.title("Least Square Regresion: Data Random")
plt.show()
```

## Interpolation

```python
from scipy.interpolate import interp1d

xi = [1, 3, 5]
yi = [20, 45, 55]

LinearInterpol = interp1d(xi, yi)

xtest = np.linspace(1, 5, 30)

ytest = LinearInterpol(xtest)

plt.scatter(xi, yi)
plt.plot(xtest, ytest)
plt.show()
```

```python
from scipy.interpolate import CubicSpline

xi = [1, 3, 5]
yi = [20, 45, 55]

CubicInterpol = CubicSpline(xi, yi)

xcase = np.linspace(1, 5, 20)
ycase = CubicInterpol(xcase)

plt.scatter(xi, yi)
plt.plot(xcase, ycase)
plt.show()
```
