## 4. Preprossessing

Intinya ini buat mastiin data yang dipake itu "optimal"

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline

from sklearn.datasets import fetch_california_housing
X, y = fetch_california_housing(return_X_y=True)
```

```python
# Pipeline = Menggabungkan preprocesssing dan model
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', KNeighborsRegressor(n_neighbors=5))
])

pipeline.fit(X, y)

pred = pipeline.predict(X)

plt.scatter(pred, y)
plt.xlabel('Predictions')
plt.ylabel('True Values')
```

Terkadang gw berpikir...

Apa yg gw lakuin ini

```python
# Pipeline = Menggabungkan preprocesssing dan model
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', KNeighborsRegressor(n_neighbors=1))
])

pipeline.fit(X, y)

pred = pipeline.predict(X)

plt.scatter(pred, y)
plt.xlabel('Predictions')
plt.ylabel('True Values')
```

## 5. Model Selection & Optimization

Untuk mencari kombinasi parameter terbaik.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
X, y = fetch_california_housing(return_X_y=True)
```

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', KNeighborsRegressor())
])

# Bikin Parameter Grid
param_grid = {
    'model__n_neighbors': [1, 3, 5, 7, 9, 11]
}

# Model Selection
model = GridSearchCV(estimator = pipe, param_grid = param_grid, cv = 5)
model.fit(X, y)

bestmodel = model.best_estimator_

pred = bestmodel.predict(X)

# Plot
plt.scatter(pred, y)
plt.xlabel('Predictions')
plt.ylabel('True Values')
plt.title(f'KNN Regression')
plt.show()
```
