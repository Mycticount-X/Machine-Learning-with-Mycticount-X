# Function Gauss-Seidel

```python
# TODO: import numpy library
import numpy as np
import matplotlib.pyplot as plt
```

```python
# TODO: buat function untuk cek diagonally dominant
def diagonal_dominant(mtx):
    # List -> Array
    mtx = np.array(mtx)
    
    # Ambil diagonal absolut
    diag = np.diag(np.abs(mtx))
    
    # Axis = 1 (Cari Row)
    # Axis = 0 (Cari Col)
    other = np.sum(np.abs(mtx), axis = 1) - diag
    
    if np.any(diag <= other):
        return False
    return True
```

```python
# Test Case 1
Xs = [
    [10, 4, 5],
    [1, 6, 1],
    [3, 1, 6]
]

Ys = [8, 5, 10]

print("Dominant:", diagonal_dominant(Xs))
print("Gauss: ")
gauss_seidel(Xs, Ys)
```

```python
# TODO: buat function gauss-seidel
def gauss_seidel(mtx, y, epsilon = 0.01, max = 15):
    if not diagonal_dominant(mtx):
        print("Not Diagonally Dominant")
        return
    
    x = np.array(mtx)
    y = np.array(y)
    
    diag = np.diag(x)
    
    x = -x
    np.fill_diagonal(x, 0)
    
    start = np.zeros(x[0].shape)
    
    for k in range(max):
        new = np.array(start)
        
        for i, r in enumerate(x):
            new[i] = (y[i] + np.dot(r, new)) / diag[i]
        
        
        print(f'Iterasi {i}:')
        print(new)
        
        dx = np.sqrt(np.dot(new - start, new - start))
        print(f'dx: {dx}')
        
        if (dx < epsilon):
            print("Converged!")
            return

        start = new
    
    print("Not Converged")
        
```
